# Analisi de la visualització 5

- ### En una grafica del tipus BoxPlot disposem de la següent informació:
    - primer quartil: el 25% dels valors son inferiors o igual a aquest valor (costat inferior de la "caixa")
    - mediana o segon quartil: el 50% dels valors son inferiors o igual a aquest valor (línia que dividex la caixa)
    - tercer quartil: el 75% dels valors son inferiors o igual a aquest valor (costat superior de la "caixa")
    - quart quartil: valor màxim (100%)
    - rang interquartílic: diferencia entre el tercer quartil i el primer quartil (mida de la "caixa": el 50% dels valors observats corresponen estan "dins" de la "caixa")
    - Bigotis: Determinen el límit per a la detecció de valors atípics. El "bigoti" tindrà una longitud màxima equivalent a 1,5 vegades el rang interquartílic. Si el límit superior o inferior no superen el 150% del rang intercuartílic, la longitud el bigoti será desde el tercer quartil fins al valor máxim i desde el primer quartil al valor mínim, respectivament.
    - Valors atípics: valors que superen el 150% del rang interquartílic, per tant queden "fora" dels bigotis.

- ### Analisi de les dades:
    - En una primera observació, podem determinar el rang en el que es troben els preus per districte i trimestre, sabent que el 50% dels valors es troben a "l'interior" de la caixa. Així doncs, podem determinar que el 50% del preus mitjans (€/mes) durant el 3 trimestre al districte de Les Corts s'an stuat entre els 1.000 i els 1.400€/mes, mentre que al districte de Sant Martí, s'han situat entre els 800 i els 1.000 €/mes durant el mateix trimestre.
    - Observem que als districtes de Ciutat Vella i Sants-Montjuic, el rang és poc ampli, el que indica que les dades estan més pròximes i per tant hi ha poca dispersió de preus.
    - En el cas del districte de les Corts, observem que la part superior de la caixa (per sobra de la mediana) és clarament superior a la part inferior, indicant així que hi ha més dispersió de dades, previsiblement per la diferencia de preus entre el barri de Pedralbes (on son molt més elevats) i la resta de barris del districte.
    - Així mateix, observem que al disricte de Sant Andreu, la dipersió es produeix per la part inferior de la mediana, en aquets cas pels preus inferiors respecte a la resta del districte, de barris com Baró de Viver o La Trinitat Vella.
    - En línies generals s'observa que a la majoria de districtes els preus han aumnetat moderadament al llarg de l'any, exceptuant al districte de Sarria Sant Gervasi, on al 2 trimestre s'ha produit una devallada del preus mitjans mensuals i tot notar-se un repunt en els preus del 4 trimestre no s'assoleixen els valors del primer tirmestre. En els cas de les Corts, observem un aument dels preus considerables des del primer al tercer trimestre, tendencia però que no es manté al 4 trimestre.

```python
# Proposa alguna visualització que creguis que pot resultar interessant.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('preu lloguer per trimestre i barri.csv')

# De la columna 'Preu' remplacem els '--' per NaN (Not a number) y pasem a 'float' els valors de la columna.

df['Preu'] = df['Preu'].replace('--',np.nan)
df['Preu'] = df['Preu'].astype(float)

# Mirem de visualitzar el preu mitjà €/mes per Districte i trimestre

    # Agafem preu del lloguer mitja mensual

df_lloguer_mensual = df.drop(df[(df['Lloguer_mitja'] == "Lloguer mitjà per superfície (Euros/m2 mes)")].index)

df_lloguer_mensual

# Fem un DF amb les columnes "Nom_Districte", "Trimestre" i la mitjana del "Preu"

df = df_lloguer_mensual.groupby(['Nom_Districte','Trimestre'])['Preu'].mean().reset_index()

# Fem un nou DF amb la funció "df.pivot" 

heatmap_df = df.pivot("Nom_Districte","Trimestre","Preu")

# Visualitzaió del heatmap

f, ax = plt.subplots(figsize=(10, 10)) # Per modificar mida del gràfic
sns.heatmap(heatmap_df, annot=True, ax=ax, fmt=".1f") # annot=True: per a que surtin els valors, fmt: per modificar visualització valor

plt.show()


```

### Comentaris

- Amb aquest gràfic, s'intenta obtenir una visió dels preus mitjans €/m2 per Districte i Trimestre, tenint en compte que les dades d'origen analitzades fan refrecnia únicmanet a l'any 2021.
- Podem observar l'evolució del Preu mitjà €/m2 per Trimetre i comparar aquests preus entre Districtes.
- Amb la variació cromatica del gràfic, podem observar els Districtes amb preus mes elevats (Sarrià-Sant Gervasi i Les Corts) i els Dsitrictes amb els preus mes baixos (Nou Barris, Sant Andreu o Horta-Guinardó)
